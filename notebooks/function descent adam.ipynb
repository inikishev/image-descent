```python
descent = FunctionDescent2D(rosenbrock)
optimizer = torch.optim.Adam(descent.parameters(), lr=2e-1)
for i in range(1000):
    optimizer.zero_grad()
    loss = descent()
    loss.backward()
    optimizer.step()
    print(i, loss, end='       \r')
descent.plot_path()
```
