# Basic tensor method
Source: https://github.com/OPTAMI/OPTAMI

```python
from stuff.found.OPTAMI.higher_order import BasicTensorMethod
from image_descent.function_descent import FunctionDescent2D
from image_descent.surfaces import *
```

```python
descent = FunctionDescent2D(rosenbrock)
optimizer = BasicTensorMethod(descent.parameters(), L = 1e-1, verbose=False)
for i in range(20):
    def closure():
        optimizer.zero_grad()
        loss = descent.step()
        #loss.backward()
        return loss
    loss = optimizer.step(closure)
    print(i, loss, end='       \r')
descent.plot_path()
```
