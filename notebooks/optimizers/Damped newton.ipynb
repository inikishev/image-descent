# Damped Newton Method
Source: https://github.com/OPTAMI/OPTAMI

```python
from stuff.found.OPTAMI.second_order import DampedNewton
from image_descent.function_descent import FunctionDescent2D
from image_descent.surfaces import *
```

```python
descent = FunctionDescent2D(rosenbrock)
optimizer = DampedNewton(descent.parameters(), L = 1e-2, verbose=False)
for i in range(100):
    def closure():
        optimizer.zero_grad()
        loss = descent.step()
        #loss.backward()
        return loss
    loss = optimizer.step(closure)
    if i%1 == 0: print(i, loss, end='       \r')
descent.plot_path()
```
