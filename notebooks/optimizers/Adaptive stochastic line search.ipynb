# Stochastic line search
Source: https://github.com/IssamLaradji/ada_sls

```python
from stuff.found.adasls import AdaSLS
from image_descent.function_descent import FunctionDescent2D
from image_descent.surfaces import *
```

```python
descent = FunctionDescent2D(rosenbrock)
optimizer = AdaSLS(descent.parameters(), n_batches_per_epoch=5, init_step_size=2)
for i in range(100):
    def closure():
        optimizer.zero_grad()
        loss = descent.step()
        #loss.backward()
        return loss
    loss = optimizer.step(closure)
    if i%10==0: print(i, loss, end='       \r')
descent.plot_path()
```
