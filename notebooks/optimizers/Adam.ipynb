```python
import torch
from image_descent.function_descent import FunctionDescent2D
from image_descent.surfaces import *
```

```python
descent = FunctionDescent2D(rosenbrock)
optimizer = torch.optim.Adam(descent.parameters(), lr=0.35)
for i in range(100):
    optimizer.zero_grad()
    loss = descent.step()
    loss.backward()
    optimizer.step()
    if i%10==0: print(i, loss, end='       \r')
descent.plot_path()
```
