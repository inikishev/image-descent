# Method of Similar Triangles
Source: https://github.com/OPTAMI/OPTAMI

```python
from stuff.found.OPTAMI.first_order import SimilarTriangles
from image_descent.function_descent import FunctionDescent2D
from image_descent.surfaces import *
```

```python
descent = FunctionDescent2D(rosenbrock)
optimizer = SimilarTriangles(descent.parameters(), L = 20, verbose=False)
for i in range(100):
    def closure():
        optimizer.zero_grad()
        loss = descent.step()
        #loss.backward()
        return loss
    loss = optimizer.step(closure)
    if i%10==0: print(i, loss, end='       \r')
descent.plot_path()
```
