# Second order random walk
Second order random walk optimizer.
On each iteration:

1. Apply random petrubation to best direction: `direction = best_direction + rand * lr`
2. Add direction to best params `params = best_params + direction`.
3. Evaluate `loss = closure(params)`
4. If loss became smaller, set `best_params` to `params`, set `best_direction` to `direction`, else generate new `direction`.

```python
from glio.optim.randomsearch import RandomWalk2
from image_descent.function_descent import FunctionDescent2D
from image_descent.surfaces import *
```

```python
descent = FunctionDescent2D(rosenbrock)
optimizer = RandomWalk2(descent.parameters(), 1e-1)
for i in range(100):
    @torch.no_grad
    def closure():
        #optimizer.zero_grad()
        loss = descent.step()
        #loss.backward()
        return loss
    loss = optimizer.step(closure)
    if i%10==0: print(i, loss, end='       \r')
descent.plot_path()
```
