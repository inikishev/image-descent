```python
from stuff.found.OPTAMI.higher_order import Hyperfast
from image_descent.function_descent import FunctionDescent2D
from image_descent.surfaces import *
```

```python
descent = FunctionDescent2D(rosenbrock)
optimizer = Hyperfast(descent.parameters(), L = 3e-1, verbose=False)
for i in range(3):
    def closure():
        optimizer.zero_grad()
        loss = descent.step()
        #loss.backward()
        return loss
    loss = optimizer.step(closure)
    if i%10==0: print(i, loss, end='       \r')
descent.plot_path()
```
