# Scipy BFGS

- https://github.com/rfeinman/pytorch-minimize
- https://github.com/scipy/scipy

```python
from torchmin.optim.scipy_minimizer import ScipyMinimizer
from image_descent.function_descent import FunctionDescent2D
from image_descent.surfaces import *
```

```python
descent = FunctionDescent2D(rosenbrock)
optimizer = ScipyMinimizer(descent.parameters(), method = 'bfgs')
for i in range(10):
    def closure():
        optimizer.zero_grad()
        loss = descent.step()
        loss.backward()
        return loss
    loss = optimizer.step(closure)
    print(i, loss, end='       \r')
descent.plot_path()
```
