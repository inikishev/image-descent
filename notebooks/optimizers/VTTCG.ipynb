# Variable three-term conjugate gradient method
source: https://github.com/kim-hansu/VTTCG

```python
from stuff.found.VTTCG import VTTCG
from image_descent.function_descent import FunctionDescent2D
from image_descent.surfaces import *
```

```python
descent = FunctionDescent2D(rosenbrock)
optimizer = VTTCG(descent.parameters(), lr=1)
for i in range(100):
    optimizer.zero_grad()
    loss = descent.step()
    loss.backward()
    optimizer.step()
    if i%10==0: print(i, loss, end='       \r')
descent.plot_path()
```
